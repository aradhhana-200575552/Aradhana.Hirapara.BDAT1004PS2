As shown above, if the file cannot be found by the interpreter or if it cannot be read
as a text file, an exception will be raised. Modify function fileLength() so that a
friendly message is printed instead:
>>> fileLength('midterm.py')
358

```python
def fileLength(filename):
    try:
        with open(filename, 'r') as infile:
            content = infile.read()
        return len(content)
    except FileNotFoundError:
        print(f"File {filename} not found.")

print(fileLength('Filelength.txt')) 
print(fileLength('idterm.py'))    # This should display the friendly error message
```

# Question 3
Write a class named Marsupial that can be used as shown below:
>>> m = Marsupial()
>>> m.put_in_pouch('doll')
>>> m.put_in_pouch('firetruck')
>>> m.put_in_pouch('kitten')
>>> m.pouch_contents()
['doll', 'firetruck', 'kitten']
Now write a class named Kangaroo as a subclass of Marsupial that inherits all the
attributes of Marsupial and also:
a. extends the Marsupial __init__ constructor to take, as input, the
coordinates x and y of the Kangaroo object,
b. supports method jump that takes number values dx and dy as input and
moves the kangaroo by dx units along the x-axis and by dy units along the yaxis, and
c. overloads the __str__ operator so it behaves as shown below.
>>> k = Kangaroo(0,0)
>>> print(k)
I am a Kangaroo located at coordinates (0,0)
>>> k.put_in_pouch('doll')
>>> k.put_in_pouch('firetruck')
>>> k.put_in_pouch('kitten')
>>> k.pouch_contents()
['doll', 'firetruck', 'kitten']
>>> k.jump(1,0)
>>> k.jump(1,0)
>>> k.jump(1,0)
>>> print(k)
I am a Kangaroo located at coordinates (3,0)

```python
class Marsupial:
    def __init__(self):
        # Initialize an empty list for the pouch
        self.pouch = []

    def put_in_pouch(self, item):
        # Add an item to the pouch list
        self.pouch.append(item)

    def pouch_contents(self):
        # Return the contents of the pouch
        return self.pouch

class Kangaroo(Marsupial):
    def __init__(self, x, y):
        # Call the superclass constructor to initialize the pouch
        super().__init__()
        # Initialize coordinates
        self.x = x
        self.y = y

    def jump(self, dx, dy):
        # Move the kangaroo by dx and dy units
        self.x += dx
        self.y += dy

    def __str__(self):
        # Override the string representation to show coordinates
        return f"I am a Kangaroo located at coordinates ({self.x},{self.y})"

# Creating a Marsupial object
m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
print(m.pouch_contents())  # Expected output: ['doll', 'firetruck', 'kitten']

# Creating a Kangaroo object
k = Kangaroo(0, 0)
print(k)  # Expected output: I am a Kangaroo located at coordinates (0,0)
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
print(k.pouch_contents())  # Expected output: ['doll', 'firetruck', 'kitten']
k.jump(1, 0)
k.jump(1, 0)
k.jump(1, 0)
print(k)  # Expected output: I am a Kangaroo located at coordinates (3,0)
```

# Question 4
Write function collatz() that takes a positive integer x as input and prints the
Collatz sequence starting at x. A Collatz sequence is obtained by repeatedly applying
this rule to the previous number x in the sequence:
x = {
𝑥/2 𝑖𝑓 𝑥 𝑖𝑠 𝑒𝑣𝑒𝑛
3𝑥 + 1 𝑖𝑓 𝑥 𝑖𝑠 𝑜𝑑𝑑
Your function should stop when the sequence gets to number 1. Your
implementation must be recursive, without any loops.
>>> collatz(1)
1
>>> collatz(10)
10
5
16
8
4
2
1

```python
def collatz(x):
    # Print the current number in the sequence
    print(x)
    # Base case: stop recursion if x is 1
    if x == 1:
        return
    # Recursive case: apply the Collatz rule
    elif x % 2 == 0:
        collatz(x // 2)  # x is even
    else:
        collatz(3 * x + 1)  # x is odd
        
collatz(1)
collatz(10)
```

# Question 5
Write a recursive method binary() that takes a non-negative
integer n and prints the binary representation of integer n.
>>> binary(0)
0
>>> binary(1)
1
>>> binary(3)
11
>>> binary(9)
1001


```python
def binary(n):
    # Base case: if n is 0 or 1, print n directly
    if n < 2:
        print(n, end='')  # end='' keeps the output on the same line
    else:
        # Recursive case: divide n by 2 and call binary() with the quotient
        binary(n // 2)
        # Print the remainder (either 0 or 1) after recursive call
        print(n % 2, end='')  # end='' keeps the output on the same line

binary(0)
print()
binary(1)
print()
binary(3)
print()
binary(9)
print()
```

# Question 6
Implement a class named HeadingParser that can be used to parse an HTML
document, and retrieve and print all the headings in the document. You should
implement your class as a subclass of HTMLParser, defined in Standard Library
module html.parser. When fed a string containing HTML code, your class should
print the headings, one per line and in the order in which they appear in the
document. Each heading should be indented as follows: an h1 heading should have
indentation 0, and h2 heading should have indentation 1, etc. Test your
implementation using w3c.html.
>>> infile = open('w3c.html')
>>> content = infile.read()
>>> infile.close()
>>> hp = HeadingParser()
>>> hp.feed(content)
W3C Mission
Principles


```python
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.indentation = 0
        self.headings = []

    def handle_starttag(self, tag, attrs):
        if tag.startswith('h') and len(tag) > 1 and tag[1].isdigit():
            self.indentation = int(tag[1]) - 1

    def handle_data(self, data):
        if data.strip() != '':
            self.headings.append((' ' * self.indentation) + data.strip())

    def print_headings(self):
        for heading in self.headings:
            print(heading)


with open('w3c.html', 'r') as f:
    content = f.read()

hp = HeadingParser()
hp.feed(content)
hp.print_headings()
```

# Question 7
Implement recursive function webdir() that takes as input: a URL (as a string) and
non-negative integers depth and indent. Your function should visit every web
page reachable from the starting URL web page in depth clicks or less, and print
each web page's URL. As shown below, indentation, specified by indent, should
be used to indicate the depth of a URL.
>>>
webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html'
, 2, 0)
http://reed.cs.depaul.edu/lperkovic/csc242/test1.html
 http://reed.cs.depaul.edu/lperkovic/csc242/test2.html
 http://reed.cs.depaul.edu/lperkovic/csc242/test4.html
 http://reed.cs.depaul.edu/lperkovic/csc242/test3.html
 http://reed.cs.depaul.edu/lperkovic/csc242/test4.html

```python
import requests
from bs4 import BeautifulSoup

def webdir(url, depth, indent):
    # Print the current URL
    print(' ' * indent + url)
    
    if depth == 0:
        return

    try:
        # Fetch page content
        response = requests.get(url)
        response.raise_for_status()  # This will raise an error for 4xx/5xx responses
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Find all the links of page
        for link in soup.find_all('a', href=True):
            # Get the absolute URL
            href = link.get('href')
            absolute_url = requests.compat.urljoin(url, href)
            
            # Recursively visit each link, with increased indent and decreased depth
            webdir(absolute_url, depth - 1, indent + 1)

    except requests.exceptions.RequestException as e:
        print(f"Could not retrieve or parse the page: {url}, error: {e}")

# Since given url was giving 404 error i have used example.com for testing purposes
webdir('https://www.example.com', 2, 0)
```

# Question 9
. Suppose list words is defined as follows:
>>> words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over',
'the', 'lazy', 'dog']
Write list comprehension expressions that use list words and generate the following
lists:
a) ['THE', 'QUICK', 'BROWN', 'FOX', 'JUMPS', 'OVER', 'THE',
'LAZY', 'DOG']
b) ['the', 'quick', 'brown', 'fox', 'jumps', 'over', 'the',
'lazy', 'dog']
c) [3, 5, 5, 3, 5, 4, 3, 4, 3] (the list of lengths of words in list
words).
d) [['THE', 'the', 3], ['QUICK', 'quick', 5], ['BROWN',
'brown', 5], ['FOX', 'fox', 3], ['JUMPS', 'jumps', 5],
['OVER', 'over', 4], ['THE', 'the', 3], ['LAZY', 'lazy',
4], ['DOG', 'dog', 3]] (the list containing a list for every word of list
words, where each list contains the word in uppercase and lowercase and the
length of the word.)
e) ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the',
'lazy', 'dog'] (the list of words in list words containing 4 or more
characters.)

```python
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']

uppercase_words = [word.upper() for word in words] # a) Convert all words to uppercase
print("a)", uppercase_words)

lowercase_words = [word.lower() for word in words] # b) Convert all words to lowercase
print("b)", lowercase_words)

word_lengths = [len(word) for word in words] # c) the list of lengths of words in list words
print("c)", word_lengths)

# d) List containing each word in uppercase, lowercase, and its length
detailed_words = [[word.upper(), word.lower(), len(word)] for word in words]
print("d)", detailed_words)

# e) List of words containing 4 or more characters
words_with_four_or_more = [word for word in words if len(word) >= 4]
print("e)", words_with_four_or_more)
```
